```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error
```

```python
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
df = pd.read_csv("/content/housing.csv", header=None, delimiter=r"\s+", names=column_names)
df
```

```python
plt.figure(figsize=(20,5))
features = ['LSTAT','RM']
target = df['MEDV']

for i,col in enumerate(features):
  plt.subplot(1, len(features), i+1)
  x = df[col]
  y = target
  plt.scatter(x,y,marker='o')
  plt.title(col)
  plt.xlabel(col)
  plt.ylabel('MEDV')
```

```python
x_rooms = df.RM
y_price = df.MEDV

x_rooms = np.array(x_rooms).reshape(-1,1)
y_price = np.array(y_price).reshape(-1,1)
print(x_rooms.shape)
print(y_price.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_rooms, y_price, test_size=0.2, random_state=5)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
reg = LinearRegression()
reg.fit(x_train,y_train)

y_train_pred = reg.predict(x_train)
rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
r2 = round(reg.score(x_train, y_train),2)
print("The model performance for training set")
print("---------------------------------------")
print("The RMSE score is ",rmse)
print("The R2 score is", r2)
```

```python
y_pred = reg.predict(x_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = round(reg.score(x_test, y_test),2)
print("The model performance for training set")
print("---------------------------------------")
print("The RMSE score is ",rmse)
print("The R2 score is", r2)
```

```python
prediction_space = np.linspace(min(x_rooms), max(x_rooms)).reshape(-1,1) 
plt.scatter(x_rooms,y_price)
plt.plot(prediction_space, reg.predict(prediction_space), color = 'black', linewidth = 3)
plt.ylabel('value of house/1000($)')
plt.xlabel('number of rooms')
plt.show()
```
