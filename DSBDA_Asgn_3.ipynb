```python
def mean(x,y):
  if y == 1:
    print(df['sepal length (cm)'].mean())
  elif y == 2:
    print(df['sepal width (cm)'].mean())
  elif y == 3:
    print(df['petal length (cm)'].mean())
  else:
    print(df['petal width (cm)'].mean())

def median(x,y):
  if y == 1:
    print(df['sepal length (cm)'].median())
  elif y == 2:
    print(df['sepal width (cm)'].median())
  elif y == 3:
    print(df['petal length (cm)'].median())
  else:
    print(df['petal width (cm)'].median())

def mode(x,y):
  if y == 1:
    print(df['sepal length (cm)'].mode())
  elif y == 2:
    print(df['sepal width (cm)'].mode())
  elif y == 3:
    print(df['petal length (cm)'].mode())
  else:
    print(df['petal width (cm)'].mode())

def std_dev(x,y):
  if y == 1:
    print(df['sepal length (cm)'].std())
  elif y == 2:
    print(df['sepal width (cm)'].std())
  elif y == 3:
    print(df['petal length (cm)'].std())
  else:
    print(df['petal width (cm)'].std())

def percentile(x,y):
  if y == 1:
    print(df['sepal length (cm)'].quantile(1))
  elif y == 2:
    print(df['sepal width (cm)'].quantile(1))
  elif y == 3:
    print(df['petal length (cm)'].quantile(1))
  else:
    print(df['petal width (cm)'].quantile(1))
```

```python
print("Available Operations are: ")
print(
    '''
    1. Mean
    2. Median
    3. Mode
    4. Standard Deviation
    5. Percentile
    '''
)
x = input("Enter the operations: ")
print(
    '''
    Available Columns are:
    1. sepal length (cm)
    2. sepal width (cm)
    3. petal length (cm)
    4. petal width (cm)
    '''
)

y = input("Enter the Column number: ")

if x == '1':
  mean(x,y)
elif x == '2':
  median(x,y)
elif x == '3':
  mode(x,y)
elif x == '4':
  std_dev(x,y)
elif x == '5':
  percentile(x,y)
```

Remove Outliers

```python
df.boxplot(column = "sepal width (cm)")
```

```python
Q1 = df.iloc[ : , : -1].quantile(0.25)
Q3 = df.iloc[ : , : -2].quantile(0.75)
IQR = Q3 - Q1
```

```python
outlier_mask = ((df.iloc[:, :-1] < (Q1 - 1.5 * IQR)) | (df.iloc[:, :-1] > (Q3 + 1.5 * IQR))).any(axis=1)
```

```python
cleaned_df = df[~outlier_mask]
```
