```python
def mean(x,y):
  if y == 1:
      print(df['fname'].mean())
  elif y == 2:
      print(df['lname'].mean())
  elif y == 3:
      print(df['marks'].mean())
  else:
      print(df['phone'].mean())

def median(x,y):
  if y == 1:
      print(df['fname'].median())
  elif y == 2:
      print(df['lname'].median())
  elif y == 3:
      print(df['marks'].median())
  else:
      print(df['phone'].median())

def mode(x,y):
  if y == 1:
      print(df['fname'].mode())
  elif y == 2:
      print(df['lname'].mode())
  elif y == 3:
      print(df['marks'].mode())
  else:
      print(df['phone'],mode())

def standard_deviation(x,y):
  if y == 1:
      print(df['fname'].std())
  elif y == 2:
      print(df['lname'].std())
  elif y == 3:
      print(df['marks'].std())
  else:
      print(df['phone'].std())

def percentile(x,y):
  if y == 1:
      print(df['fname'].quantile(1))
  elif y == 2:
      print(df['lname'].quantile(1))
  elif y == 3:
      print(df['marks'].quantile(1))
  else:
      print(df['phone'].quantile(1))
```

```python
print("Available Operations are: ")
print('''
    1. Mean
    2. Median
    3. Mode
    4. Standard Deviation
    5. Percentile
''')
x = input('Enter the Operations: ')
print('''
    Available Columns are:
    1. fname
    2. lname
    3. marks
    4. phone
''')
y = input('Enter the Columns: ')

if x == '1':
     mean(x,y)
elif x == '2':
      median(x,y)
elif x == '3':
      mode(x,y)
elif x == '4':
      standard_deviation(x,y)
elif x == '5':
      percentile(x,y)
```
